```python
import sys
import matplotlib
matplotlib.use('Agg')

import matplotlib.pyplot as plt 
import numpy as np
x = np.array(["A","B","C","D"])
y = np.array([3,8,1,10])

plt.bar(x,y)
plt.show()


sys.stdout.flush()
```

```python
#Three lines to make our compiler able to draw:
import sys
import matplotlib
matplotlib.use('Agg')

import matplotlib.pyplot as plt
import numpy as np

x = ["APPLES", "BANANAS"]
y = [400, 350]

plt.bar(x, y)
plt.show()

#Two  lines to make our compiler able to draw:
sys.stdout.flush()
```

```python
#horizontal graph
#Three lines to make our compiler able to draw:
import sys
import matplotlib
matplotlib.use('Agg')

import matplotlib.pyplot as plt
import numpy as np

x = ["APPLES", "BANANAS"]
y = [400, 350]

plt.barh(x, y) #"BARH"
plt.show()

#Two  lines to make our compiler able to draw:
sys.stdout.flush()
```

```python
#color bar graph
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color = "#ff4000")
plt.show()
```

```python
#bar width
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, width = 0.1)
plt.show()
```

```python
#pie chart
import matplotlib.pyplot as plt
import numpy as np

y = np.array([70, 69, 61]) 

plt.pie(y)
plt.show() 
```

```python
#pie chart labels
import matplotlib.pyplot as plt
import numpy as np

y = np.array([70, 69, 61]) 
mylabels = ["single variable", "several variable", "Chemistry"]
plt.pie(y, labels = mylabels)
plt.show() 
```

```python
#colors pie chart
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 55, 54, 45])
mylabels = ["bEEE", "single variable", "several variable", "Chemistry"]
mycolors = ["#ff4000", "#1affff", "#99ff33", "#4CAF50"]

plt.pie(y, labels = mylabels, colors = mycolors)
plt.show() 
```

```python
#colors pie chart LEGEND
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 55, 54, 45])
mylabels = ["bEEE", "single variable", "several variable", "Chemistry"]
mycolors = ["#ff4000", "#1affff", "#99ff33", "#4CAF50"]

plt.pie(y, labels = mylabels, colors = mycolors)
plt.legend()
plt.show() 
```
